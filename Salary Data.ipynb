```python
#Preproccessing the data before implementing ML ALOGRITHM
```

```python
df = df.loc[:, ['workclass',"fnlwgt","education","education-num","occupation",
                   "capital-gain","capital-loss","hours-per-week","native-country","salary"]]
```

```python
df.head()
```

```python
#now using the label encoder
#changing the values to numeric variable
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab_enc = LabelEncoder()

df['workclass'] = lab_enc.fit_transform(df['workclass'])
df['education'] = lab_enc.fit_transform(df['education'])
df['occupation'] = lab_enc.fit_transform(df['occupation'])
df['native-country'] = lab_enc.fit_transform(df['native-country'])
df['salary'] = lab_enc.fit_transform(df['salary'])
```

```python
df.head()
```

```python
Now using the ML alogrithms
```

```python
X = df.iloc[:, 0:-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
```

```python
from sklearn.ensemble import RandomForestClassifier

regressor =RandomForestClassifier(n_estimators=200, random_state=0)  
regressor.fit(X_train, y_train)  
y_pred = regressor.predict(X_test) 
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix 

cm = confusion_matrix(y_test, y_pred)
print(cm)

accuracy = accuracy_score(y_test, y_pred)*100
print('Accuracy of our model is equal ' + str(round(accuracy, 2)) + ' %.')
```
